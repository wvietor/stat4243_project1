```python
import anndata
import scanpy as sc
```

# Read the cleaned data and normalize by count

```python
ad = anndata.read_h5ad('cleaned_processed_frogtail.h5ad')
sc.pp.normalize_total(ad, target_sum=1e4)
sc.pp.neighbors(ad, metric='cosine')
```

# Exploratory Analysis

## Cluster-Wise Correlation Matrix

```python
sc.pl.correlation_matrix(ad, 'cluster')
```

It's clear that the erythrocyte cells are quite distinct from all the others but similar to each other.

## UMAP (cosine distance)

```python
sc.tl.umap(ad)
sc.pl.umap(ad, color='cluster')
```

## t-SNE (cosine distance)

```python
sc.tl.tsne(ad)
sc.pl.tsne(ad, color='cluster')
```

```python
sc.tl.rank_genes_groups(ad, groupby='cluster', method='logreg')
sc.pl.rank_genes_groups_heatmap(ad, n_genes=15, groupby='cluster')
```
